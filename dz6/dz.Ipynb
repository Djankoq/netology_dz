```python
import pandas as pd
```

```python
vist_log = pd.read_csv('visit_log.csv', decimal=',')
vist_log.dropna(subset=['user_id'])
vist_log.sort_values("user_id")
vist_log = pd.DataFrame(vist_log)
vist_log['user_id'] = vist_log['user_id'].astype(str)
vist_log
```

```python
purchase_log = pd.read_csv('purchase_log.txt', decimal=',')
purchase_log.dropna(subset=['user_id'])
purchase_log = pd.DataFrame(purchase_log.apply(lambda x: x.str.split(':').str[-1]))
purchase_log = purchase_log.map(lambda x: x.replace('"', ''))
purchase_log = purchase_log.map(lambda x: x.replace('}', ''))
purchase_log
```

```python
vist_log['user_id'] = vist_log['user_id'].str.strip()
purchase_log['user_id'] = purchase_log['user_id'].str.strip()
```

```python
new_table = pd.merge(vist_log, purchase_log, on='user_id', how='inner')
new_table
new_table.to_csv('funnel.csv', index=False)
```
